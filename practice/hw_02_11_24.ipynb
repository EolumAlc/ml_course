# Task description

The present dataset (trojan_detection.csv) contains the malicious and benign network packets which can be used to detect Trojan horse. The dataset contains 100758 rows and 86 columns. The first 85 columns are the features and the last column is the label. The dataset is imbalanced with 90683 malicious and 10075 benign packets. The dataset is available at [this Kaggle link](https://www.kaggle.com/datasets/kurasbedianigrigoriy/trojan-detection-modified).

The task is binary classification. The goal is to detect Trojan horse in the network packets.

## EDA

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from seaborn import kdeplot
```

```python
# Load data

df = pd.read_csv('../datasets/trojan_detection.csv')
df.head()
```

```python
# Clean column names

df.columns = df.columns.map(lambda x: x.strip())
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('/', '_')
```

```python
# Drop id column

df.drop('id', axis=1, inplace=True)
```

```python
# Encode target variable

from sklearn.preprocessing import LabelEncoder

# rename Class to target
df.rename(columns={'class': 'target'}, inplace=True)
df['target'] = LabelEncoder().fit_transform(df['target'])
```

```python
# Get info about the data

df.info()
```

```python
# Get summary statistics

df.describe()
```

```python
# Unique values in the columns

df.nunique().sort_values().reset_index().rename(columns={'index': 'Column', 0: 'Number unique values'})
```

```python
# Drop columns with only one unique value

df.drop(df.columns[df.nunique() == 1], axis=1, inplace=True)
```

```python
# Check missing values

df.isna().sum().reset_index().rename(columns={'index': 'Column', 0: 'Number of missing values'})
```

```python
# Find numerical and categorical features

numerical_columns = df.select_dtypes(exclude=['object']).columns
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_columns, categorical_columns
```

```python
# Remove target from numerical columns

numerical_columns = numerical_columns.drop('target')
```

```python
# Statistics for numerical features

df[numerical_columns].describe()
```

```python
# Fill missing values in numerical features with mode

df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mode().iloc[0])
```

```python
# Statistics for categorical features

df[categorical_columns].describe()
```

```python
# Check unique values in the categorical features

df[categorical_columns].nunique().reset_index().rename(columns={'index': 'Column', 0: 'Number of unique values'})
```

```python
# Fill missing values in categorical features with mode

df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
# Encode categorical features
## Encode IP
import ipaddress

df.source_ip = df.source_ip.apply(lambda x: np.float64(int(ipaddress.IPv4Address(x))))
df.destination_ip = df.destination_ip.apply(lambda x: np.float64(int(ipaddress.IPv4Address(x))))
```

```python
# Drop Flow ID and Timestamp columns

df.drop('flow_id', axis=1, inplace=True)
df.drop('timestamp', axis=1, inplace=True)
```

```python
# Normalize features

from sklearn.preprocessing import StandardScaler

df[numerical_columns] = StandardScaler().fit_transform(df[numerical_columns])
```

```python
# Plot target variable 

norm_target = (df
               .target
               .value_counts(normalize=True)
               .mul(100)
               .rename('percent')
               .reset_index())
plt.figure(figsize=(15, 7))
sns.barplot(x='target', y='percent', data=norm_target)
plt.title('Target variable', fontsize=20)
plt.xlabel('Target', fontsize=14)
plt.ylabel('Percent', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# Plot distribution histograms

df.hist(figsize=(20, 20), bins=50)
plt.show()
```

```python
# Split data and target

X = df.drop('target', axis=1)
y = df.target.astype(np.int64)
```

```python
# Check correlation between numerical features and the target variable

X.corrwith(y).rename('Correlation with target').sort_values(ascending=False)
```

```python
# Plot correlation between features

plt.figure(figsize=(20, 20))
sns.heatmap(X.corr(), annot=False, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Find columns where correlation with the target variable is greater than threshold

threshold = 0.02
columns_where_correlation_with_target_is_greater_than_threshold = X.corrwith(y).index[np.abs(X.corrwith(y)) > threshold]
columns_where_correlation_with_target_is_greater_than_threshold
```

```python
# Find correlated features and drop them,
# because they can lead to multicollinearity

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

correlation_matrix = X.corr().abs()
threshold = 0.9
high_corr_features = [
    (col1, col2) for col1 in correlation_matrix.columns for col2 in correlation_matrix.columns
    if col1 != col2 and correlation_matrix.loc[col1, col2] > threshold
]
features_to_drop = set()
for col1, col2 in high_corr_features:
    if col1 not in features_to_drop and col2 not in features_to_drop:
        features_to_drop.add(col2)
X = X.drop(columns=features_to_drop)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf_model.fit(X, y)
feature_importances = rf_model.feature_importances_
selector = SelectFromModel(rf_model, threshold='mean', prefit=True)
selected_features = X.columns[selector.get_support()]
X = X[selected_features]
```

## KNN

```python
# Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Apply kNN

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [75, 80, 85],
    'weights': ['distance'],
    'metric': ['euclidean']
}

grid = GridSearchCV(knn, param_grid, cv=5, scoring='f1', n_jobs=-1)

grid.fit(X_train, y_train)
```

```python
grid.best_params_

# {'metric': 'manhattan', 'n_neighbors': 15, 'weights': 'uniform'}  f1=0.9488147681980281
# {'metric': 'euclidean', 'n_neighbors': 40, 'weights': 'distance'} f1=0.9492156759876186
# {'metric': 'euclidean', 'n_neighbors': 60, 'weights': 'distance'} f1=0.9492751723595564
# {'metric': 'euclidean', 'n_neighbors': 80, 'weights': 'distance'} f1=0.9493408810965224
```

```python
# Apply kNN with best parameters

# knn = KNeighborsClassifier(**grid.best_params_)
knn = KNeighborsClassifier(**{'metric': 'euclidean', 'n_neighbors': 80, 'weights': 'distance'})
knn.fit(X_train, y_train)
```

```python
# Apply kNN with best parameters on test set

y_pred = knn.predict(X_test)
```

```python
# Check metrics
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay


def check_metrics(clf, y_true, y_pred, tittle=''):
    print(classification_report(y_true, y_pred, zero_division=0, digits=5))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Blues, cbar=False,
                xticklabels=clf.classes_,
                yticklabels=clf.classes_)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix' + tittle)
    plt.show()


check_metrics(knn, y_test, y_pred)
```

Based on the conditions of the problem, the preferred option is when the amount of FP is minimal.
