# Task description

The present dataset (trojan_detection.csv) contains the malicious and benign network packets which can be used to detect Trojan horse. The dataset contains 100758 rows and 86 columns. The first 85 columns are the features and the last column is the label. The dataset is imbalanced with 90683 malicious and 10075 benign packets. The dataset is available at [this Kaggle link](https://www.kaggle.com/datasets/kurasbedianigrigoriy/trojan-detection-modified).

The task is binary classification. The goal is to detect Trojan horse in the network packets.

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
# Load data

df = pd.read_csv('../datasets/trojan_detection.csv')
df.head()
```

```python
# Clean column names

df.columns = [i.strip() for i in df.columns]
```

```python
# Get info about the data

df.info()
```

```python
# Get summary statistics

df.describe()
```

```python
# Drop columns with only one unique value

df.drop(df.columns[df.nunique() == 1], axis=1, inplace=True)
```

```python
# Drop id column

df.drop('id', axis=1, inplace=True)
```

```python
# Find numerical and categorical features

numerical_columns = df.select_dtypes(exclude=['object']).columns
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_columns, categorical_columns
```

```python
# Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2) 

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Check for missing values

X_train.isna().sum()
```

```python
# Make pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', MinMaxScaler())
])

X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
```

```python
# Apply kNN

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [1, 3, 5, 7, 9, 11, 13, 15, 17, 25, 30, 40, 50, 60, 7080, 90, 100],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid = GridSearchCV(knn, param_grid, cv=5, scoring='f1', n_jobs=-1)

# grid.fit(X_train, y_train)
```

```python
# grid.best_params_

# {'metric': 'manhattan', 'n_neighbors': 15, 'weights': 'uniform'}  f1=0.9488147681980281
# {'metric': 'euclidean', 'n_neighbors': 40, 'weights': 'distance'} f1=0.9492156759876186
# {'metric': 'euclidean', 'n_neighbors': 60, 'weights': 'distance'} f1=0.9492751723595564
# {'metric': 'euclidean', 'n_neighbors': 80, 'weights': 'distance'} f1=0.9493408810965224
```

```python
# Apply kNN with best parameters

# knn = KNeighborsClassifier(**grid.best_params_)
knn = KNeighborsClassifier(**{'metric': 'euclidean', 'n_neighbors': 80, 'weights': 'distance'})
knn.fit(X_train, y_train)
```

```python
# Apply kNN with best parameters on test set

y_pred = knn.predict(X_test)
```

```python
# Check metrics
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

def check_metrics(clf, y_true, y_pred, tittle=''):
    print(classification_report(y_true, y_pred, zero_division=0, digits=5))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Blues, cbar=False,
                xticklabels=clf.classes_,
                yticklabels=clf.classes_)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix' + tittle)
    plt.show()
            
        


check_metrics(knn, y_test, y_pred)
```

Based on the conditions of the problem, the preferred option is when the amount of FP is minimal.

## Logistic Regression

```python
# Apply GridSearchCV

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {'penalty': ['l1', 'l2'],
              'C': np.logspace(-3, 3, 7),
              'solver': ['liblinear', 'lbfgs'],
              'max_iter': [1000],
              'n_jobs': [-1],
              'random_state': [42]
}

gs = GridSearchCV(LogisticRegression(),
                  param_grid,
                  cv=5,
                  scoring='f1',
                  verbose=2,
                  n_jobs=-1)

gs.fit(X_train, y_train)
gs.best_params_
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(**gs.best_params_)
log_reg.fit(X_train, y_train)
log_reg.score(X_train, y_train)
```

```python
# Metrics for test set
check_metrics(log_reg, y_test, log_reg.predict(X_test), tittle=' (test set)')
```

```python
# Metrics for train set
check_metrics(log_reg, y_train, log_reg.predict(X_train), tittle=' (train set)')
```

When using logistic regression, the problem of weights being shifted towards class 1 is clearly visible. This is due to the imbalance of classes.

## Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier

param_dist = {'n_estimators': range(10, 301, 50),
              'max_depth': range(10, 101, 10),
}
rs = RandomizedSearchCV(RandomForestClassifier(),
                        param_dist,
                        cv=5,
                        scoring='f1',
                        verbose=2,
                        n_jobs=-1,
                        random_state=42)
rs.fit(X_train, y_train)
rs.best_params_
```

```python
rs_df = pd.DataFrame(rs.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
rs_df = rs_df.drop([
            'mean_fit_time', 
            'std_fit_time', 
            'mean_score_time',
            'std_score_time', 
            'params', 
            'split0_test_score', 
            'split1_test_score', 
            'split2_test_score', 
            'std_test_score'],
            axis=1)
rs_df
```

```python
# Visualize the results of RandomizedSearchCV

plt.figure(figsize=(10, 10))
sns.scatterplot(data=rs_df, x='param_n_estimators', y='param_max_depth', hue='mean_test_score', size='mean_test_score', sizes=(50, 500))
plt.show()
```

```python
# Apply GridSearchCV

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {'n_estimators': [10, 160],
              'max_depth': [10, 60, 80, 100],
}

gs = GridSearchCV(RandomForestClassifier(),
                  param_grid,
                  cv=5,
                  scoring='f1',
                  verbose=2,
                  n_jobs=-1)

gs.fit(X_train, y_train)
gs.best_params_
```

```python
gs_df = pd.DataFrame(gs.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
gs_df = gs_df.drop([
            'mean_fit_time', 
            'std_fit_time', 
            'mean_score_time',
            'std_score_time', 
            'params', 
            'split0_test_score', 
            'split1_test_score', 
            'split2_test_score', 
            'std_test_score'],
            axis=1)
gs_df
```

```python
gs.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42, **gs.best_params_)
rf.fit(X_train, y_train)
rf.score(X_train, y_train)
```

```python
# Metrics for test set
y_pred = rf.predict(X_test)
check_metrics(rf, y_test, y_pred, tittle=' (test set)')
```

```python
# Metrics for train set
check_metrics(rf, y_train, rf.predict(X_train), tittle=' (train set)')
```

Random forest showed better quality, but there is still a problem associated with class imbalance.

## MLP

```python
# Apply GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

param_grid_mlp = {
    'hidden_layer_sizes': [(i, j) for i in range(50, 101, 50) for j in range(50, 101, 50)],
    'alpha': [0.0001, 0.0005],

}

grid_search_mlp = GridSearchCV(estimator=MLPClassifier(max_iter=1000), param_grid=param_grid_mlp, cv=5, scoring='f1', n_jobs=-1, verbose=2)

grid_search_mlp.fit(X_train, y_train)
grid_search_mlp.best_params_
```

```python
grid_search_mlp_df = pd.DataFrame(grid_search_mlp.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
grid_search_mlp_df = grid_search_mlp_df.drop([
            'mean_fit_time', 
            'std_fit_time', 
            'mean_score_time',
            'std_score_time', 
            'params', 
            'split0_test_score', 
            'split1_test_score', 
            'split2_test_score', 
            'std_test_score'],
            axis=1)
grid_search_mlp_df
```

```python
# Apply MLP with best parameters

from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(max_iter=1000, random_state=42, **grid_search_mlp.best_params_)
mlp.fit(X_train, y_train)
mlp.score(X_train, y_train)
```

```python
# Metrics for test set
y_pred = mlp.predict(X_test)
check_metrics(mlp, y_test, y_pred, tittle=' (test set)')
```

```python
# Metrics for train set
check_metrics(mlp, y_train, mlp.predict(X_train), tittle=' (train set)')
```

Neither underfitting nor overfitting is observed

```python
# Check metrics for all models

from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

models = [knn, log_reg, rf, mlp]
models_names = ['kNN', 'Logistic Regression', 'Random Forest', 'MLP']
metrics_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1'])
for i, model in enumerate(models):
    y_pred = model.predict(X_test)
    metrics_df.loc[i] = [models_names[i], accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred), f1_score(y_test, y_pred)]
metrics_df.sort_values('F1', ascending=False)
```

# Report

For our task, the correct identification of Trojan packets is critically important, rather than the incorrect identification of Benign ones. Therefore, we can assume that all methods showed good results except logistic regression, which blocks all network packets.

All methods have a problem with class imbalance. To solve this, can use class weighting or oversampling/undersampling.
